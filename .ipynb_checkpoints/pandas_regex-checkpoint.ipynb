This notebook reproduces examples in https://kanoki.org/2019/11/12/how-to-use-regex-in-pandas/
<p>Except we user data from 2021 instead on 2019</p>

```python
import pandas as pd
# df = pd.read_csv('./world-happiness-report-2019.csv')
df = pd.read_csv('../input/world-happiness-report/world-happiness-report-2021.csv')
```

```python
df.head()
```

```python
import numpy as np
df['first_five_Letter']=df['Country name'].str.extract(r'(^w{5})')
df.head()
```

```python
S=pd.Series(['Finland','Colombia','Florida','Japan','Puerto Rico','Russia','france'])
S[S.str.count(r'(^F.*)')==1]
```

```python
# Total items starting with F
S.str.count(r'(^F.*)').sum()
```

```python
df[df['Country name'].str.count('^[pP].*')>0]
```

```python
df[df['Country name'].str.match('^[UC].*')== True]
```

```python
# Remove the dash(-) followed by number from all countries in the Series
S=pd.Series(['Finland-1','Colombia-2','Florida-3','Japan-4','Puerto Rico-5','Russia-6','france-7'])
S.replace('(-d)','',regex=True, inplace = True)
```

```python
S
```

```python
S=pd.Series(['Finland','Colombia','Florida','Japan','Puerto Rico','Russia','france'])
[itm[0] for itm in S.str.findall('^[Ff].*') if len(itm)>0]
```

```python
S=pd.Series(['Finland','Colombia','Florida','Japan','Puerto Rico','Russia','france'])
S.str.contains('^F.*')
```

```python
df[df['Country name'].str.contains('^I.*')==True]
```

```python
s = pd.Series(["StatueofLiberty built-on 28-Oct-1886"])
s.str.split(r"s", n=-1,expand=True)
```
